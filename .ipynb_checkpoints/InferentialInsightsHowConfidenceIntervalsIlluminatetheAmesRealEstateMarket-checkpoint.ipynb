```python
import scipy.stats as stats
import pandas as pd

Ames = pd.read_csv('Ames.csv')

#Define the confidence level and degrees of freedom
confidence_level = 0.95
degrees_freedom = Ames['SalePrice'].count()- 1

#Calculate the confidence interval for 'SalePrice'
confidence_interval = stats.t.interval(confidence_level, degrees_freedom, loc=Ames['SalePrice'].mean(), scale=Ames['SalePrice'].sem())

# Print out the sentence with the confidence interval figures
print(f"The 95% confidence interval for the true mean sales price of all houses in Ames is between ${confidence_interval[0]:.2f} and ${confidence_interval[1]:.2f}.")
```

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Cargar datos
Ames = pd.read_csv('Ames.csv')

# Parmetros del intervalo de confianza
confidence_level = 0.95
degrees_freedom = Ames['SalePrice'].count() - 1

# Calcular el intervalo de confianza
confidence_interval = stats.t.interval(
    confidence_level,
    degrees_freedom,
    loc=Ames['SalePrice'].mean(),
    scale=Ames['SalePrice'].sem()
)

# Crear histograma
plt.figure(figsize=(10, 7))
plt.hist(Ames['SalePrice'], bins=30, color='lightblue', edgecolor='black', alpha=0.5,
         label='Sales Prices Distribution')

# Lneas verticales para la media y el intervalo de confianza
plt.axvline(Ames['SalePrice'].mean(), color='blue', linestyle='-',
            label=f'Mean: ${Ames["SalePrice"].mean():,.2f}')
plt.axvline(confidence_interval[0], color='red', linestyle='--',
            label=f'Lower 95% CI: ${confidence_interval[0]:,.2f}')
plt.axvline(confidence_interval[1], color='green', linestyle='--',
            label=f'Upper 95% CI: ${confidence_interval[1]:,.2f}')

# Etiquetas y estilo
plt.title('Distribution of Sales Prices with Confidence Interval', fontsize=20)
plt.xlabel('Sales Price', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.xlim([min(Ames['SalePrice']) - 5000, max(Ames['SalePrice']) + 5000])
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Cargar los datos
Ames = pd.read_csv('Ames.csv')

# Definir nivel de confianza y grados de libertad
confidence_level = 0.95
degrees_freedom = Ames['SalePrice'].count() - 1

# Calcular el intervalo de confianza para 'SalePrice'
confidence_interval = stats.t.interval(
    confidence_level,
    degrees_freedom,
    loc=Ames['SalePrice'].mean(),
    scale=Ames['SalePrice'].sem()
)

# Crear la figura centrada en la media y el intervalo de confianza
plt.figure(figsize=(10, 7))
plt.hist(Ames['SalePrice'], bins=30, color='lightblue', edgecolor='black', alpha=0.5,
         label='Sales Prices')

# Enfocar el eje x alrededor de la media
plt.xlim([150000, 200000])

# Lneas verticales para la media y los lmites del intervalo
plt.axvline(Ames['SalePrice'].mean(), color='blue', linestyle='-',
            label=f'Mean: ${Ames["SalePrice"].mean():,.2f}')
plt.axvline(confidence_interval[0], color='red', linestyle='--',
            label=f'Lower 95% CI: ${confidence_interval[0]:,.2f}')
plt.axvline(confidence_interval[1], color='green', linestyle='--',
            label=f'Upper 95% CI: ${confidence_interval[1]:,.2f}')

# Etiquetas y leyenda
plt.title('Zoomed-in View of Mean and Confidence Intervals', fontsize=20)
plt.xlabel('Sales Price', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.legend()
plt.grid(axis='y')
plt.show()
```
