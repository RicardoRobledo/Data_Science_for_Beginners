```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
Ames = pd.read_csv('Ames.csv')

# Calcular la correlación de todas las variables con 'SalePrice'
correlations = Ames.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)

# Obtener las 5 variables más correlacionadas con 'SalePrice' (excluyendo 'SalePrice' en sí)
top_5_features = correlations.index[1:6]

# Crear un DataFrame solo con esas variables + 'SalePrice'
selected_features = Ames[top_5_features.tolist() + ['SalePrice']]

# Crear un pair plot para visualizar relaciones entre estas variables
sns.pairplot(selected_features, height=2.5, aspect=1.2)

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
Ames = pd.read_csv('Ames.csv')

# Convert 'LotShape' to a binary feature: 'Regular' and 'Irregular'
Ames['LotShape_Binary'] = Ames['LotShape'].apply(lambda x: 'Regular' if x == 'Reg' else 'Irregular')

# Creating the pair plot, color-coded by 'LotShape_Binary'
sns.pairplot(Ames, vars=['SalePrice', 'OverallQual', 'GrLivArea'], hue='LotShape_Binary', palette='Set1', height=2.5, aspect=1.75)

# Display the plot
plt.show()
```
